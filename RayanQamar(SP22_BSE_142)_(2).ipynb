```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [7, 9, 15, 22, 28, 33, 31, 30, 27, 21, 15, 9],
    [10, 12, 18, 26, 33, 37, 36, 35, 31, 25, 19, 12],
    [18, 19, 22, 26, 29, 32, 31, 31, 29, 27, 23, 20],
    [5, 7, 12, 18, 24, 28, 27, 26, 23, 17, 11, 6]
])

monthly_avg = np.mean(data, axis=0)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10, 6))
plt.plot(months, data[0], label='Islamabad', marker='o')
plt.plot(months, data[1], label='Lahore', marker='s')
plt.plot(months, data[2], label='Karachi', marker='^')
plt.plot(months, data[3], label='Quetta', marker='d')

plt.title('Monthly Average Temperatures for Cities in Pakistan', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Temperature (C)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(months, monthly_avg, color='skyblue')

plt.title('Average Monthly Temperature Across All Cities in Pakistan', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Temperature (C)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def hill_height(x):
    return -1 * (x - 5)**2 + 20

def climb_hill(start_position):
    position = start_position
    path = [position]

    while True:
        current_height = hill_height(position)
        left_height = hill_height(position - 1)
        right_height = hill_height(position + 1)

        if left_height > current_height and left_height > right_height:
            position -= 1
        elif right_height > current_height and right_height > left_height:
            position += 1
        else:
            break

        path.append(position)

    return path

start_position = 0
path = climb_hill(start_position)


x_values = np.linspace(0, 10, 100)
y_values = hill_height(x_values)

plt.plot(x_values, y_values, color="black")
plt.scatter(path, hill_height(np.array(path)), color="red")
plt.title("Climbing the Hill")
plt.xlabel("x")
plt.ylabel("Height")
plt.legend()
plt.grid(True)
plt.show()
```

Task 1: NumPy Array Creation

```python
import numpy as np

arr1 = np.arange(10)
print("1D array (0 to 9):", arr1)
```

1.2.1 Create a 2D array with shape (3, 3), where values range from 1 to 9

```python
arr2 = np.arange(1, 10).reshape(3, 3)
print("2D array (1 to 9):\n", arr2)
```

1.3.1 Create a 3D array of shape (2, 3, 4) filled with random integers from 1 to 100

```python
arr3 = np.random.randint(1, 101, size=(2, 3, 4))
print("3D array (random integers 1 to 100):\n", arr3)
```

Task 1: Basic Line Plot
1.1.1 Create a basic line plot:

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

Task 2: Scatter Plot
1.2.1 Create a scatter plot:

```python
# Data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, s=100, c='blue', alpha=0.7, edgecolors='black')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

Task 3: Bar Plot
1.3.1 Create a bar plot

```python
# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Plot
plt.figure(figsize=(8, 5))
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.bar(categories, values, color=colors)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Task 4: Histogram
1.4.1 Generate a histogram:

```python
import numpy as np

# Generate random numbers
data = np.random.randn(1000)

# Plot
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Task 5: Pie Chart
1.5.1 Create a pie chart:

```python
# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]
explode = (0.1, 0, 0, 0)  # explode the 1st slice (Python)

# Plot
plt.figure(figsize=(8, 5))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Programming Language Popularity')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges=[
    ("Islamabad","Rawalpindi"),
    ("Islamabad","Lahore"),
    ("Islamabad","Peshawar"),
    ("Rawalpindi","Peshawar"),
    ("Rawalpindi","Quetta"),
    ("Peshawar","Quetta"),
    ("Lahore","Multan"),
    ("Lahore","Quetta"),
    ("Multan","Quetta"),
    ("Multan","Karachi"),
    ("Quetta","Karachi")
]
G.add_edges_from(edges)
def bfs_shortest_path(graph,start_city,end_city):
 return nx.shortest_path(graph, start_city, end_city)

start="Islamabad"
end="Karachi"
shortest_path=bfs_shortest_path(G,start,end)

pos=nx.spring_layout(G)
plt.figure(figsize=(10,6))
nx.draw(G, pos, with_labels=True,node_color="lightblue", node_size=3000, font_size=10, font_weight="bold",edge_color="gray")

patch_edges=list(zip(shortest_path[:-1],shortest_path[1:]))
nx.draw_networkx_edges(G,pos,edgelist=patch_edges,edge_color="red",width=3)


plt.show()

print(f"Shortest path from {start} to {end}: {shortest_path}")
```
